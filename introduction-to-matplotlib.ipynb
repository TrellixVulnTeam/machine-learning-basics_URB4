## matplotlib example workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt

x = [-4,-3,-2,-1,0,1,2,3,4]
y= [16,9,4,1,0,1,4,9,16]

fig, ax = plt.subplots(figsize=(10,10))

ax.plot(x,y)

ax.set(title ="Simple Plot",
      xlabel = "x-axis",
      ylabel = "y-axis")
fig.savefig("sample-plot.png")
```

#  oo method


```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# pyplot method
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            figsize=(10,6));
```

```python
#  oo method
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x= "age",
            y="chol",
            c='target',
            ax=ax);
ax.set_xlim([45,80]);
```

```python
# OO method from scratch

fig, ax = plt.subplots(figsize = (10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

# Customize the plot
ax.set(title="Heart Disease and Cholestrol Levels",
      xlabel="Age",
      ylabel="cholestrol")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target")

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
over_50.head()
```

```python
# OO method from scratch

fig, (ax1, ax2) = plt.subplots(figsize = (10,10),
                            nrows=2,
                            ncols=1,
                            sharex=True)

# Plot the data to ax1
scatter = ax1.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

# Customize the plot of ax1
ax1.set(title="Heart Disease and Cholestrol Levels",
#       xlabel="Age",
      ylabel="Cholestrol")

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a horizontal line to ax1
ax1.axhline(over_50["chol"].mean(),
          linestyle="-.",
           c="red");

# Plot the data to ax2
scatter = ax2.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"])

# Customize the plot to ax2
ax2.set(title="Heart Disease and Thalach Levels",
      ylabel="Thalach",
      xlabel="Age")

# Add a legend to ax2
ax2.legend(*scatter.legend_elements(), title="Target")

# Add a horizontal line to ax2
ax2.axhline(over_50["thalach"].mean(),
          linestyle="-.",
          animated = True,
          c="red");

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
